```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset desde un archivo CSV (ajusta la ruta si es necesario)
df = pd.read_csv("dataset_sintetico.csv")   # Cambia esto por el nombre correcto del archivo
```

```python
# Mostrar las primeras filas del dataset
print("Primeras filas del dataset:")
print(df.head())
```

```python
# Información general del dataset
print("\nInformación del dataset:")
print(df.info())
```

```python
# Estadísticas descriptivas de las variables numéricas
print("\nEstadísticas descriptivas:")
print(df.describe())
```

```python
# Seleccionar solo columnas numéricas para los histogramas
columnas_numericas = df.select_dtypes(include=['number']).columns
```

```python
# Crear histogramas de variables numéricas
df[columnas_numericas].hist(figsize=(10, 8), bins=30, edgecolor='black')
plt.suptitle("Distribución de Variables Numéricas", fontsize=16)
plt.show()
```
